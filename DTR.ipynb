```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading data

```python
from sklearn.datasets import fetch_california_housing
df = fetch_california_housing()
```

```python
#this is targeted value it's not in input the housing of pricee
df = fetch_california_housing()
# Access the target values (median house value)
target_values = df.target  # Use 'data.target' instead of 'data['target']'

# Display the first few target values
print(target_values[:10])  # Display the first 10 target values
```

```python
df.feature_names
```

```python
data = pd.DataFrame(df.data , columns = df.feature_names) #convert into fdata frame
```

```python
data.head()
```

```python
# Add a column for the target (median house value)
data['MedHouseVal'] = df.target
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

## spliting the data

```python
x = data.drop(['MedHouseVal'], axis=1)   #input
```

```python
y = data['MedHouseVal'] #output
```

```python
x.head()
```

```python
y.head()
```

### ascaling our data ranges to have comman range for all the features

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
```

```python
x = sc_x.fit_transform(x) #transform input data
```

### splitiing test train data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3 , random_state = 42)
```

```python

x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

# DECIASION TREE REGRESSION

```python
from sklearn.tree import DecisionTreeRegressor 
dtr = DecisionTreeRegressor()
```

```python
#train model

dtr.fit(x_train, y_train)
```

```python
#predict 
y_pred = dtr.predict(x_test)
```

```python
y_pred[:10]
```

```python
y_test
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_pred, y_test)
```

```python
# Create a scatter plot with seaborn
sns.scatterplot(x=y_test, y=y_pred)

# Add labels and title
plt.xlabel('Actual Values (y_test)')
plt.ylabel('Predicted Values (y_pred)')
plt.title('Actual vs Predicted Values')

# Plot a diagonal line for reference (perfect predictions)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linewidth=2)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Plot using the first feature of x_test
sns.scatterplot(x=x_test[:, 0], y=y_test, color='blue', label='Actual Values (y_test)')
sns.scatterplot(x=x_test[:, 0], y=y_pred, color='red', label='Predicted Values (y_pred)')

# Add labels and title
plt.xlabel('First Feature of x_test')
plt.ylabel('Target Values')
plt.title('Actual vs Predicted Values')

# Show the plot with legend
plt.legend()
plt.show()
```
